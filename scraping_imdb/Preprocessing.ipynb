```python
import pandas as pd
```

```python
df = pd.read_csv (r'movies1000.csv')
df
```

```python
df['metascore']=df['metascore'].str.extract('(\d+)')
```

```python
df['release_date']=df['release_date'].str.extract('(\d+)')
```

```python
df['genre']=df['genre'].str.replace('[', '', regex=True).replace(']', '', regex=True).replace(',', '|', regex=True).replace(r"'", '', regex=True).replace(r"  ", '', regex=True)
```

```python
df['movieStars']=df['movieStars'].str.replace('[', '', regex=True).replace(']', '', regex=True).replace(',', '|', regex=True).replace(r"'", '', regex=True).replace(r" ", '', regex=True)
```

```python
df['movieDirector']=df['movieDirector'].str.replace('Directors:', '', regex=True).replace(',', '|', regex=True)
```

```python
df.to_csv('Preprocessingmovies1000.csv', index=False)
```
